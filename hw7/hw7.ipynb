## task 4

Рассмотрим задачу для Коши дифференциального уравнения на функцию u(t)

$$
\frac {du}
{dt} = \frac 1 u
$$
c начальным условием u(t = 0) = 1 на интервале 0 ≤ t ≤ 5.
Составить в явном виде расчетные формулы неявного методом Эйлера для данной задачи. Найти порядок
аппроксимации схемы. Используя схему с постоянным шагом τ = 0.1, построить численное решение.
Сравнить с точным решением задачи Коши.

```python
T = 5
dt = 0.1
u0 = 1

num_steps = int(T/dt)
tt = np.arange(num_steps+1)*dt  # сетка
y = np.empty(num_steps+1)
y[0] = u0
for k in range(num_steps):
    y[k+1] = 0.5 * (y[k] + np.sqrt(y[k]**2 + 4*dt))

true_y = np.sqrt(2*tt + 1)
```

```python
plt.plot(tt, y, '-', label='Euler')
plt.plot(tt, true_y, '-', label='exact')

plt.legend()
plt.show()
```
